```python
wb = Workbook(write_only=True)
ws = wb.create_sheet()

driver = webdriver.Chrome()
driver.get("https://www.youtube.com/watch?v=4CEvMrPw5xw")
link = "https://www.youtube.com/watch?v=4CEvMrPw5xw"
driver.implicitly_wait(3)

time.sleep(3)

driver.execute_script("window.scrollTo(0,800)")
time.sleep(3)

#driver = webdriver.Chrome("")

last_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(1.5)

    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
time.sleep(1.5)
# 팝업 닫기 시도
try:
    driver.find_elements(By.CSS_SELECTOR, "#dismiss-button > a").click()
except:
    #더보기 클릭
    pass
    buttons = driver.find_elements(By.CSS_SELECTOR, "#more-replies > a")
    time.sleep(1.5)

    for button in buttons:
        button.send_keys(Keys.ENTER)
        time.sleep(1.5)
        button.click()
    
    #페이지 소스 가져오기
    html_source = driver.page_source
    soup = BeautifulSoup(html_source,'html.parser')
    
    #댓글과 제목 가져오기
    comment_list = soup.select("#content-text > span")
    title_list = soup.select("#title > h1")
    
    #담을거 가져오기
    comment_final = []
    title_final = []
    link_final = []
    thumbnail_url_final = []
    
    # 제목이 있을 경우 첫 번째 제목을 가져오고, 없으면 빈 문자열로 설정
    if title_list:
        temp_title = title_list[0].text  # 첫 번째 제목을 가져옴
        temp_title = temp_title.replace('\n','')
        print(f"제목이 제대로 추출되었습니다: {temp_title}")  # 제목 확인용 출력
    else:
        temp_title = ""  # 제목이 없을 경우 빈 문자열로 설정
        print("제목을 찾을 수 없습니다.")  # 제목이 없을 때의 메시지
    #썸네일 URL 추출
    video_id = link.split('=')[1]
    thumbnail_url = f"https://img.youtube.com/vi/{video_id}/maxresdefault.jpg"
    
    for i in range(len(comment_list)):
        temp_comment = comment_list[i].text
        temp_comment = temp_comment.replace('\n', '')
        temp_comment = temp_comment.replace('\t', '')
        temp_comment = temp_comment.replace('    ', '')
        comment_final.append(temp_comment) # 댓글 내용
        title_final.append(temp_title) # 제목
        link_final.append(link) #링크
        thumbnail_url_final.append(thumbnail_url)
# DataFrame 생성
pd_data = {
    "제목":title_final, 
#    "댓글내용": comment_final,
    "링크":link_final,
    "썸네일":thumbnail_url_final
}
nyoutube_pd = pd.DataFrame(pd_data)
#youtube_pd.to_excel('댓글저장.xlsx')
nyoutube_pd
```

```python
# 제목별로 댓글내용을 합치고 링크는 중복 없이 하나만 유지
def combine_group(group):
    combined_comments = ' '.join(group['댓글내용'])
    unique_links = group['링크'].unique()
    combined_links = ' '.join(unique_links)
    unique_thumbnail = group['썸네일URL'].unique()
    combined_thumbnails = ''.join(unique_thumbnail)
    return pd.Series({'댓글내용': combined_comments, '링크': combined_links,'썸네일 URL': combined_thumbnails})
```

```python
# 그룹화하여 제목별로 댓글과 링크를 합침
result_df = nyoutube_pd.groupby('제목').apply(combine_group2).reset_index()
```
